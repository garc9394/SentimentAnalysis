```python
# Dependencies

import tweepy
import json
import apikeys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Twitter API key and authentication

consumer_key = apikeys.TWITTER_CONSUMER_KEY
consumer_secret = apikeys.TWITTER_CONSUMER_SECRET
access_token = apikeys.TWITTER_ACCESS_TOKEN
access_token_secret = apikeys.TWITTER_ACCESS_TOKEN_SECRET

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# News organizations (Twitter users instead of search terms such as #BBCNews)

newsOrg = ("BBCNews", "CBSNews", "CNN", "FoxNews", "nytimes")
```

```python
# Create an empty dict

sentiment_data = {'org':[],
                 'tweets_ago':[],
                 'text':[],
                 'compound':[],
                 'positive':[],
                 'negative':[],
                 'neutral':[]}

# Call for the recent 100 tweets for each of the news organizations; analyze tweets with Vader Sentiment

for org in newsOrg:
    tweets_ago = 0
    public_tweets = api.user_timeline(org, count=100, result_type="recent")
    for tweet in public_tweets:
        text = tweet['text']
        scores = analyzer.polarity_scores(text)
        compound = scores['compound']
        pos = scores['pos']
        neg = scores['neg']
        neu = scores['neu']
        tweets_ago += 1

# Append sentiments to dict

        sentiment_data['org'].append(org)
        sentiment_data['tweets_ago'].append(tweets_ago)
        sentiment_data['text'].append(text)
        sentiment_data['compound'].append(compound)
        sentiment_data['positive'].append(pos)
        sentiment_data['negative'].append(neg)
        sentiment_data['neutral'].append(neu)
```

```python
# Create DataFrame from dict

sentiment_data_df = pd.DataFrame(sentiment_data)
sentiment_data_df

sentiment_data_df.to_csv('sentiment_data.csv', sep='\t', encoding='utf-8')
```

```python
# Use "org" as index

sentiment_data_df = sentiment_data_df.set_index("org")
sentiment_data_df
```

```python
# Create DataFrame for each of the news organizations

bbc_df = sentiment_data_df.loc["BBCNews"]
bbc_df
```

```python
cbs_df = sentiment_data_df.loc["CBSNews"]
cbs_df
```

```python
cnn_df = sentiment_data_df.loc["CNN"]
cnn_df
```

```python
fox_df = sentiment_data_df.loc["FoxNews"]
fox_df
```

```python
nytimes_df = sentiment_data_df.loc["nytimes"]
nytimes_df
```

```python
# Today's date

date_of_analysis = datetime.datetime.now().strftime('%m-%d-%y')

# Values for each of the news organizations

bbc_xvalues = bbc_df['tweets_ago']
bbc_yvalues = bbc_df['compound']
cbs_xvalues = cbs_df['tweets_ago']
cbs_yvalues = cbs_df['compound']
cnn_xvalues = cnn_df['tweets_ago']
cnn_yvalues = cnn_df['compound']
fox_xvalues = fox_df['tweets_ago']
fox_yvalues = fox_df['compound']
nytimes_xvalues = nytimes_df['tweets_ago']
nytimes_yvalues = nytimes_df['compound']

plt.figure(figsize=(10, 8), dpi=80)
bbc = plt.scatter(bbc_xvalues, bbc_yvalues, marker="o",
            facecolors="lightblue", edgecolors="black",
            s=80)
cbs = plt.scatter(cbs_xvalues, cbs_yvalues, marker="o",
            facecolors="green", edgecolors="black",
            s=80)
cnn = plt.scatter(cnn_xvalues, cnn_yvalues, marker="o",
            facecolors="red", edgecolors="black",
            s=80)
fox = plt.scatter(fox_xvalues, fox_yvalues, marker="o",
            facecolors="blue", edgecolors="black",
            s=80)
nytimes = plt.scatter(nytimes_xvalues, nytimes_yvalues, marker="o",
            facecolors="yellow", edgecolors="black",
            s=80)
title = ("Sentiment Analysis of Media Tweets (" + str(date_of_analysis) + ")")
plt.title(title)
plt.xlabel("Tweets Ago")
plt.ylabel("Tweet Polarity")
plt.xlim(105,-19) # In descending order
plt.ylim(-1.15,1.15)
plt.grid(True)

plt.legend((bbc,cbs,cnn,fox,nytimes),("BBC","CBS","CNN","Fox","NY Times"))

plt.show()

plt.savefig("Images/Sentiment_Analysis_of_Media_Tweets.png")
```

```python
# Calculate mean of compound sentiments for each of the news organizations

bbc_compound = bbc_df['compound'].mean()
cbs_compound = cbs_df['compound'].mean()
cnn_compound = cnn_df['compound'].mean()
fox_compound = fox_df['compound'].mean()
nytimes_compound = nytimes_df['compound'].mean()

# X-axis labels

orgs = ["BBC", "CBS", "CNN", "Fox", "NY Times"]
compounds = [bbc_compound, cbs_compound, cnn_compound, fox_compound, nytimes_compound]
x_values = np.arange(len(orgs))
colors = ['lightblue', 'green', 'red', 'blue', 'yellow']
```

```python
# Add bar chart options, and plot chart

plt.figure(figsize=(8, 6), dpi=80)
ax = plt.subplot()
ax.set_facecolor("lightgrey")
plt.xlim(-0.5, 4.5)
plt.bar(x_values, compounds, color=colors, alpha=0.5,
        tick_label=orgs, width=1, edgecolor='black')
plt.title("Overall Media Sentiment of Tweets (" + str(date_of_analysis) + ")")
plt.ylabel("Tweet Polarity")
# add "compounds" value as bar label

plt.show()
plt.savefig("Images/Overall_Media_Sentiment_of_Tweets.png")
```
